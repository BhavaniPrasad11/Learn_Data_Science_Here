```python
# Create a pivot table using the 'survived' column as values, 'sex' as index, and 'class' as columns
df_titanic.pivot_table('survived', index = 'sex', columns = 'class')
```

```python
# Create age bins and pivot the DataFrame 'df_titanic' with 'survived' as values, 'sex' and age bins as index, and 'class' as columns
age = pd.cut(df_titanic['age'], [0, 18, 80])
df_titanic.pivot_table('survived', ['sex', age], 'class')
```

```python
# Create fare bins and pivot the DataFrame 'df_titanic' with 'survived' as values, 'sex' and age bins as index, and fare bins and 'class' as columns
fare = pd.qcut(df_titanic['fare'], 2)
df_titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
```

```python
df_titanic.pivot_table(index = 'sex', columns = 'class', aggfunc = {'survived': sum, 'fare': 'mean'})
```

```python
# Create a pivot table using the 'survived' column as values, 'sex' as index, 'class' as columns, and include margins
df_titanic.pivot_table('survived', index = 'sex', columns = 'class', margins = True)
```

```python
df_Glo = pd.read_csv('Global YouTube Statistics.csv', encoding = 'latin-1')
df_Glo.head(3)
```

```python
non_num_features = [feature for feature in df_Glo.columns if df_Glo[feature].dtype == 'object']
num_features = [feature for feature in df_Glo.columns if df_Glo[feature].dtype != 'object']
```

```python
df_num = df_Glo[num_features]
df_num
```

```python
# Calculate the number of unique values in each column of the DataFrame 'df_Glo'.
df_Glo.nunique()
```

```python
# Calculate the correlation matrix for the columns in the DataFrame 'df_num'.
df_num.corr()
```

```python
plt.figure(figsize = (15, 8))
plt.bar(height = 'subscribers', x = 'Youtuber', data = df_Glo.head(15))
plt.xticks(rotation = 90)
plt.title('Top 15 Youtubers in terms of Subscribers')
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(15, 15))

# Create a mask for the upper triangle of the correlation matrix
msk = np.triu(np.ones_like(df_num.corr()))

# Create a heatmap with annotations, using a colormap, and applying the mask
sns.heatmap(df_num.corr(), annot=True, cmap=plt.cm.RdBu_r, mask=msk)

# Display the heatmap
plt.show()
```

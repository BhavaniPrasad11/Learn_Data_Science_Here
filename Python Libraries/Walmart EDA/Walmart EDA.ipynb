* convert date column into date time format

```python
df['Order Date'] = pd.to_datetime(df['Order Date'] , format = '%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'] , format = '%d-%m-%Y')
```

* seperate day, month & year

```python
df['O_Day'] = df['Order Date'].dt.day
df['O_Month'] = df['Order Date'].dt.month
df['O_Year'] = df['Order Date'].dt.year
```

```python
df.head()
```

```python
def quarter(month):
  if month in [1,2,3]:
    return 'Q1'
  elif month in [4,5,6]:
    return 'Q2'
  elif month in [7,8,9]:
    return 'Q3'
  else:
    return 'Q4'
```

```python
#Use this function to identify the quarter of prder placed
df['Quarter'] = df['O_Month'].apply(quarter)
```

```python
df.head()
```

* Calculate the duration company took to ship a product

```python
df['Ship Date'] - df['Order Date']
```

```python
df['Processing Time'] = (df['Ship Date'] - df['Order Date']).apply(lambda x : x.days)
```

```python
import numpy as np

table = pd.pivot_table(data = df,
                       index = 'Category',                              #grouping factor
                       values = ['Order ID' , 'Profit', 	'Quantity' 	,'Sales'],      #Targeted columns
                       aggfunc = {'Order ID': pd.Series.nunique,
                                  'Profit' : np.mean,
                                  'Quantity' : np.sum ,
                                  'Sales' : np.mean})                                           #statistical calculations

table.columns = ["Total Orders" , "Average Profit" , "Total Quantity" , "Average Sales"]

table = table.sort_values('Total Quantity' , ascending = False)
table
```

* Create a column representing profit or loss
* scatter plot representing sales vs profit , green color for profit & red color for loss

```python
df['P/L'] = df['Profit'].apply(lambda x : 'g' if x > 0 else 'y' if x == 0 else 'r' )
```

```python
df.head()
```

```python
df['P/L'].value_counts()
```

```python
plt.figure(figsize = (12,8))

plt.scatter(df.Sales , df.Profit , c = df['P/L'] , marker = '*')

plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize = (12,8))

plt.scatter(df.O_Year , df.Profit , c = df['P/L'] , marker = '*')

plt.xlabel('Year')
plt.ylabel('Profit')
plt.show()
```

```python
df[['Sales','Profit','P/L']].plot.scatter(x = 'Sales' , y = 'Profit' , c = 'P/L')
```

```python
explode = [0.1 , 0 , 0]
df['P/L'].value_counts().plot.pie(autopct='%1.1f%%' , explode = explode , shadow = True ,
                                  labels = ['Profit' , 'Loss' , 'Breakeven'])
```

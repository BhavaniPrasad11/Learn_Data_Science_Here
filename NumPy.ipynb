```python
A = np.array([[3,4,1] , [2,2,2] , [4,1,1]])
print(A)
X = np.array([['x'] , ['y'] , ['z']])
print(X)
Z = np.array([[7] , [14] , [21]])
print(Z)
```

```python
#A.X = Z
#X = invA.Z

invA = np.linalg.inv(A)
X = np.dot(invA , Z)
print(X)
```
